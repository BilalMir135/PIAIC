# Our Model

```python
def line(x, w=0, b=0):
    y = w*x+b #y=mx+b
    return y
```

# Cost Function

```python
def mean_squared_error(y_true, y_pred):
    s = (y_true-y_pred)**2
    return s.mean()
```

```python
X = df[['Height']].values
y_true = df[['Weight']].values
```

```python
y_pred = line(X) #for now W=0 and b=0
```

```python
df.plot(kind='scatter', x='Height', y='Weight', title='Weight and Height in adults')
plt.plot([55,78],[75,250], color='yellow', linewidth=5)

plt.plot(X,y_pred, color='red', linewidth=5)
```

```python
mean_squared_error(y_true,y_pred)
```

```python
#adjusting weights #1
y_pred = line(X,1)

df.plot(kind='scatter', x='Height', y='Weight', title='Weight and Height in adults')
plt.plot([55,78],[75,250], color='yellow', linewidth=5)

plt.plot(X,y_pred, color='red', linewidth=5)
```

```python
mean_squared_error(y_true,y_pred)
```

```python
#adjusting weights #2
y_pred = line(X,4,5)

df.plot(kind='scatter', x='Height', y='Weight', title='Weight and Height in adults')
plt.plot([55,78],[75,250], color='yellow', linewidth=5)

plt.plot(X,y_pred, color='red', linewidth=5)
```

```python
mean_squared_error(y_true,y_pred)
```

Manually it is very difficult to adjust the weights. In this case we have only one weight but when training model with images, when there are hundreds or thousands of weights it will become impossible for us to adjust all of them but model do these adjustments in a seconds

```python
#adjusting weights by model
y_pred = line(X,7.8,-347)

df.plot(kind='scatter', x='Height', y='Weight', title='Weight and Height in adults')
plt.plot([55,78],[75,250], color='yellow', linewidth=5)

plt.plot(X,y_pred, color='red', linewidth=5)
```
