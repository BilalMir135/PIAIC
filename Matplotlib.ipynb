# Bar Diagram

```python
x = [1990,1991,1992,1993,1994,1995]
y = [20,22,35,50,60,80]
plt.bar(x,y)
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales of a Store')
plt.show()
```

# Multiple Bar Diagram

```python
import numpy as np
y1=[8,10,12]
y2=[18,20,32]
y3=[28,32,48]
x=np.arange(len(y1))
bar_width=0.2  #Increasing bar_width so the bar do not overlap each other
plt.bar(x,y1,label='First',width=0.3,zorder=2)   #zorder = 2 so the bar comes front of grid lines
plt.bar(x+bar_width,y2,label='Second',width=0.3,zorder=2)
plt.bar(x+bar_width*2,y3,label='Third',width=0.3,zorder=2)
plt.xticks(x+bar_width,['1996','1997','1998']) 
plt.xlabel('Years')
plt.ylabel('Number of Students')
plt.title('Students Division')
plt.legend()
plt.grid(axis='y')
plt.show()
```

# Stack Plot

```python
days = [1,2,3,4,5]

eating=[2,3,4,3,2]
sleeping=[10,8,8,12,6]
working=[5,10,8,5,2]
playing=[7,1,4,4,14]

plt.stackplot(days,eating,sleeping,working,playing)  #we cannot use legends in stackplot
plt.xlabel('Days')
plt.ylabel('Time')
plt.title('Stack Plot')
plt.show()
```

```python
#One solution of legends in stackplot
days = [1,2,3,4,5]

eating=[2,3,4,3,2]
sleeping=[10,8,8,12,6]
working=[5,10,8,5,2]
playing=[7,1,4,4,14]

plt.plot([],[], label='Eating', color='r',linewidth=5)
plt.plot([],[], label='Sleeping', color='orange',linewidth=5)
plt.plot([],[], label='Working', color='g',linewidth=5)
plt.plot([],[], label='Playing', color='b',linewidth=5)

plt.stackplot(days,eating,sleeping,working,playing,colors=['r','orange','g','b'])
plt.xlabel('Days')
plt.ylabel('Time')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

# Pie Chart

```python
expenditure = [1200,1000,500,300]
items = ['Food','House Rent','Clothing','Education']
plt.pie(expenditure,labels=items)
plt.title('Pie Chart')
plt.legend()
plt.show()
```

```python
expenditure = [1200,1000,500,300]
items = ['Food','House Rent','Clothing','Education']
plt.pie(expenditure,labels=items,startangle=90)  #Giving our own starting angle
plt.title('Pie Chart')
plt.show()
```

```python
expenditure = [1200,1000,500,300]
items = ['Food','House Rent','Clothing','Education']
plt.pie(expenditure,labels=items,startangle=90,shadow=True)  #Giving shadow
plt.title('Pie Chart')
plt.show()
```

```python
expenditure = [1200,1000,500,300]
items = ['Food','House Rent','Clothing','Education']
plt.pie(expenditure,labels=items,startangle=90,shadow=True,explode=(0,0,0,0.2))  #Exploding out the slice of education
plt.title('Pie Chart')
plt.show()
```

```python
expenditure = [1200,1000,500,300]
items = ['Food','House Rent','Clothing','Education']
plt.pie(expenditure,labels=items,startangle=90,shadow=True,explode=(0,0,0,0.2),autopct='%1.1f%%') #showing percentage of each slice
plt.title('Pie Chart')
plt.show()
```

# Plotting Data stored in any file

```python
import csv

x=[]
y=[]
with open('sample.txt','r') as file:
    data = csv.reader(file,delimiter=',')
    for row in data:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph')
plt.show()
```

```python
#using numpy to load data
import numpy as np
x,y = np.loadtxt('sample.txt',delimiter=',',unpack=True)     #unpack will automatically load the , separated values of row separately in x qand y
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph')
plt.show()
```
