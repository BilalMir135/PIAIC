```python
import tensorflow as tf
```

# Loading Data

```python
data = tf.keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels) = data.load_data()
```

```python
train_images = train_images/255
test_images = test_images/255
```

# Building Model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),  #.Flatten() is used to converty data in vetor(1d tensor)
                                   tf.keras.layers.Dense(512, activation=tf.nn.relu), #nn stands for neural network
                                   tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

# Compiling Model

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# Training Model

```python
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

# Testing Model

```python
test_acc = model.evaluate(test_images,test_labels)
```
